# CoShop Data Explorer

This notebook demonstrates loading and exploring the CoShop benchmark datasets.

Datasets available: `movielens` (movies), `goodreads` (books), `hm` (clothing)

```python
from coshop.data import get_dataset, DATASETS
print("Available datasets:", DATASETS)
```

## MovieLens (movies)

```python
ml = get_dataset('movielens')
print(f"Dataset: movielens | Item type: {ml[0].item_name} | Catalog size: {len(ml.catalog)}")
ml.catalog[['title', 'year', 'genres', 'vote_average']].head(5)
```

```python
# A Specification is a single task: a user's preferences and a ground-truth target set
spec = ml[0]
print("Spec index:", 0)
print("Item name:", spec.item_name)
print("xstar (ground truth target IDs):", spec.xstar[:5], "...")
print("zstar (ground truth preferences):", spec.zstar)
```

## Goodreads (books)

```python
gr = get_dataset('goodreads')
print(f"Dataset: goodreads | Item type: {gr[0].item_name} | Catalog size: {len(gr.catalog)}")
gr.catalog[['title', 'author_name', 'genres']].head(5)
```

## H&M (clothing)

```python
hm = get_dataset('hm')
print(f"Dataset: hm | Item type: {hm[0].item_name} | Catalog size: {len(hm.catalog)}")
hm.catalog.head(5)
```

## Utility Function Demo

Each specification has a `column_ustar` utility function that scores catalog items against the ground-truth preferences.

```python
spec = ml[0]
# Score the first 10 catalog items
sample_ids = list(ml.catalog.index[:10])
if hasattr(spec, 'column_ustar') and spec.column_ustar is not None:
    scores = spec.column_ustar(sample_ids)
    import pandas as pd
    results = pd.DataFrame({'item_id': sample_ids, 'ustar_score': scores})
    results = results.merge(
        ml.catalog[['title']].reset_index(),
        left_on='item_id', right_on='id'
    )
    print(results[['title', 'ustar_score']].sort_values('ustar_score', ascending=False).to_string())
else:
    print("No column_ustar available for this spec")
```
